```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("cardio_train.csv", sep=";")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df = df.drop('id', axis=1)
df.head()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
# transforming age from days to year
df['age'] = (df['age']/365).round().astype(int)
df.head()
```

## Handling Outliers

```python
print(df['age'].min())
print(df['age'].max())
```

```python
print(df['height'].min())
print(df['height'].max())

sns.scatterplot(df['height'])
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.show()
```

```python
print(len(df[df['height'] < 150]))
print(len(df[df['height'] > 200]))
```

```python
len(df[df['height'] < 140])
```

```python
df = df[df['height'] > 140]
```

```python
len(df[df['height'] < 150])
```

```python
df.iloc[df['height'] < 150] = 150
len(df[df['height'] < 150])
```

```python
print(df['weight'].min())
print(df['weight'].max())

sns.scatterplot(df['weight'])
plt.title('Distribution of weight')
plt.xlabel('Weight')
plt.ylabel('Frequency / Density')
plt.show()
```

```python
len(df[df['weight'] <= 40])
```

```python
len(df[df['weight'] <= 30])
```

```python
df = df[df['weight'] >= 30]
```

```python
df.iloc[df['weight'] < 40] = 40
len(df[df['weight'] < 40])
```

```python
print(df['ap_hi'].min())
print(df['ap_hi'].max())

print(len(df[df['ap_hi'] < 90]))
print(len(df[df['ap_hi'] > 240]))
```

```python
opposite = df[df["ap_hi"] < df["ap_lo"]]
print(len(opposite))
opposite.head()
```

```python
swap = df['ap_hi'] < df['ap_lo']
df.loc[swap, ['ap_hi', 'ap_lo']] = df.loc[swap, ['ap_lo', 'ap_hi']].values
```

```python
len(df[df["ap_hi"] < df["ap_lo"]])
```

```python
df.shape
```

```python
print(len(df[df['ap_hi'] < 90]))
print(len(df[df['ap_hi'] > 240]))

print(len(df[df['ap_lo'] < 60]))
print(len(df[df['ap_lo'] > 160]))
```

```python
df = df[df["ap_hi"].between(90, 240)]
df = df[df["ap_lo"].between(60, 160)]
df.shape
```
